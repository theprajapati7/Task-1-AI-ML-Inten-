### Handle Numerical Columns

```python
df['age'].fillna(df['age'].mean(), inplace=True)
df['annual_income'].fillna(df['annual_income'].median(), inplace=True)
df['purchase_amount'].fillna(df['purchase_amount'].mean(), inplace=True)
```

### Handle Categorical Columns

```python
df['gender'].fillna(df['gender'].mode()[0], inplace=True)
df['city'].fillna(df['city'].mode()[0], inplace=True)
```

### Date Columns

```python
df['signup_date'].fillna('Unknown', inplace=True)
```

```python
df
```

### Label Encoding (for Binary Categories)

```python
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1})
```

```python
df
```

### One-Hot Encoding (for Multiple Categories)

```python
df = pd.get_dummies(df, columns=['city'], drop_first=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['age', 'annual_income', 'purchase_amount']] = scaler.fit_transform(df[['age', 'annual_income', 'purchase_amount']])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['age', 'annual_income', 'purchase_amount']] = scaler.fit_transform(df[['age', 'annual_income', 'purchase_amount']])
```

###  Visualize Outliers Using Boxplots

```python
# Plot boxplot for 'age', 'annual_income', and 'purchase_amount'
sns.boxplot(x=df['age'])
plt.show()

sns.boxplot(x=df['annual_income'])
plt.show()

sns.boxplot(x=df['purchase_amount'])
plt.show()
```

### Removing Outliers

```python

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Remove rows with outliers
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Apply the function to remove outliers
df = remove_outliers(df, 'age')
df = remove_outliers(df, 'annual_income')
df = remove_outliers(df, 'purchase_amount')
```

###  Verify the Changes

```python
sns.boxplot(x=df['age'])
plt.show()

sns.boxplot(x=df['annual_income'])
plt.show()

sns.boxplot(x=df['purchase_amount'])
plt.show()
```
